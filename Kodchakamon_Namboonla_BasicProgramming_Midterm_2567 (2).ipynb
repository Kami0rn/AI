สเปกตรัมของแสง โดยปกติสามารถบรรยายได้ด้วยฟังก์ชั่น Lorentzian หรือฟังก์ชั่น Gaussian ตามสมการ

$$y_\rm{Lorentzian} = A \cdot \left(\frac{\Gamma^2}{(x - x_0)^2 + \Gamma^2} \right) $$
$$y_\rm{Gaussian} = A \cdot \rm{exp} \left( - \frac{(x - x_0)^2}{\sigma^2} \right) $$

เราจะมาดูว่าฟังก์ชันใดสอดคล้องกับสเปกตรัมของเรามากกว่า โดยการพยายามฟิตและคำนวณหาค่าคงเหลือ (residue)

## ข้อ 3 (10 คะแนน)

จงเขียนฟังก์ชั่นชื่อ `Lorentzian` ที่รับ parameter 4 ตัวตามลำดับ คือ
1. ลิสต์ของค่า x
2. ค่า A
3. ค่า $x_0$
3. ค่า $\Gamma$
และให้ผลลัพธ์ return ออกมาเป็นลิสต์ของค่าฟังก์ชั่น Lorentzian ที่สอดคล้องกับลิสต์ของค่า x

และฟังก์ชั่นชื่อ `Gaussian` ที่รับ parameter 4 ตัวตามลำดับ คือ
1. ลิสต์ของค่า x
2. ค่า A
3. ค่า $x_0$
3. ค่า $\sigma$
และให้ผลลัพธ์ return ออกมาเป็นลิสต์ของค่าฟังก์ชั่น Gaussian ที่สอดคล้องกับลิสต์ของค่า x

แนะ: สามารถใช้ np.array() ในการเปลี่ยนลิสต์เป็น numpy array ได้

```python
# เขียนโค้ด
def Lorentzian(xs, A, x_0, gamma):
    loren_0 = []
    for i in xs:
        y_loren = A * (gamma**2 /( (( i - x_0)**2) + (gamma**2)))
        loren_0.append(y_loren)
    return np.array(loren_0)
A = 1 #เช็ค
gamma = 2 #เช็ค
x_0 = xs[0] #เช็ค
print(Lorentzian(xs, A, x_0, gamma)) #เช็ค

def Gaussian(xs, A, x_0, sigma):
    gauss_0 = []
    for i in xs:
        y_gauss = A * np.exp((-(( i - x_0)**2))/(sigma**2))
        gauss_0.append(y_gauss)
    return np.array(gauss_0)
A = 4 #เช็ค
sigma = 2 #เช็ค
x_0 = xs[0] #เช็ค
print(Lorentzian(xs, A, x_0, sigma)) #เช็ค
```

## ข้อ 6 (10 คะแนน, ใช้ for loop เท่านั้น ห้ามใช้ numpy)

จงคำนวณค่าผลต่างของฟังก์ชั่นที่ได้เทียบกับข้อมูล $y_\rm{residue} = y_\rm{Lorentzian} - y_\rm{data}$ โดยใช้ for loop ลบเลขสมาชิกแต่ละตัว สร้างออกมาเป็นลิสต์ใหมชื่อ `y_res` และพลอตกราฟของ `y_res`

```python
# เขียนโค้ด
y_res = []
for i in range(len(x_data)):
    yi = ys_lorentzian_fig[i] - y_data[i]
    y_res.append(yi)
print(y_res)
plt.plot(x_data, y_res, '-', color = 'b', label = 'residue')
```

## ข้อ 7 (10 คะแนน, ใช้ for loop เท่านั้น ห้ามใช้ numpy)
การหาความคลาดเคลื่อน e ของการเดานี้ สามารถหาได้จากผลรวมของผลต่างกำลังสองระหว่างค่าที่เดาได้และค่าที่ได้จากการทดลองจริง (sum of squared difference)

$$ e = \sum_i (y_{\rm{Lorenzian}, i} - y_{\rm{data}, i})^2 $$

จงหาค่า $e$ จากลิสต์ของ `y_res` ในข้อ 6 โดยการเขียน for loop (ห้ามใช้ numpy ในการคำนวณ)

```python
# เขียนโค้ด
e = 0
for i in range(len(y_res)):
    e += (ys_lorentzian_fig[i] - y_data[i])**2
print(e)
```

## ข้อ 8 (15 คะแนน)
8.1 เขียนฟังก์ชั่น `calc_error(xs, A, x0, Gamma)` โดยที่ฟังก์ชันนี้จะต้อง return ค่าความคลาดเคลื่อน `e` ออกมา เมื่อป้อนค่า `Gamma` เข้าไป

```python
# เขียนโค้ด
def calc_error(xs, A, x_0, gamma):
    ys_loren_fig = Lorentzian(xs, A, x_0, gamma)
    e = 0
    for i in range(len(xs)):
        e += (ys_loren_fig[i] - y_data[i])**2
    return e
gamma = float(input("Please enter Gamma : "))
print("Error e:", calc_error(xs, A, x_0, gamma))
```

8.2 สร้างค่า `Gamma` หลายๆ ค่าในช่วงที่เหมาะสม และเขียนโปรแกรมหาค่า `Gamma` จากลิสต์นี้ที่ทำให้ค่า `e` มีค่าน้อยที่สุด

(ไม่ต้องเปลี่ยนค่า $A$ และ $x_0$) จากข้อ 5

```python
# เขียนโค้ด
Gamma = np.linspace(100, 1000, 10000)
error_0 = calc_error(xs, 0.145, 650, 200)
best_gamma = 0
min_error = error_0
for i in Gamma:
    error = calc_error(xs, 0.145, 650, i)
    if error < min_error:
        min_error = error
        best_gamma = i
print(f'The best gamma is: {best_gamma:.2f} with a minimum error of: {min_error:.6f}')
```

## ข้อ 9 (15 คะแนน)

เปลี่ยนค่าทั้งสาม (`A`, `x0`, `Gamma`) ในช่วงที่เหมาะสมและรายงานค่า `A`, `x0`, `Gamma` ที่ดีที่สุดสำหรับข้อมูลที่ได้

```python
# เขียนโค้ด รันโค้ดนานมาก37นาทีค่ะ
A_lst = np.linspace(0.1, 0.2, 10)
x_0_lst = np.linspace(500, 700, 201)
Gamma_lst = np.linspace(10, 1000, 10000)
error_0_1 = calc_error(x_data, 0.145, 650, 670)
best_A = 0
best_x_0= 0
best_Gamma = 0
min_error_0 = error_0_1
for a in A_lst:
    for x in x_0_lst:
        for g in Gamma_lst:
            error = calc_error(x_data, a, x, g)
            if error < min_error_0:
                min_error_0 = error
                best_A = a
                best_x_0 = x
                best_Gamma = g
print("best A : ", best_A  )
print("best x0 : ", best_x_0  )
print("best gamma : ", best_Gamma  )
print("minimum error : ", min_error_0  )
```
